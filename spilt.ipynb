```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('NCAA_Tourney_2002-2017.csv')
```

```python
df.rename(columns=lambda x: x.replace('team1', 'team3'), inplace=True)
```

```python
df.rename(columns=lambda x:x.replace('team2','team1'),inplace=True)
```

```python
df.rename(columns=lambda x:x.replace('team3','team2'),inplace=True)
```

```python
df['result'] = pd.Series(np.where(df['result'] == 1, 0, 1),df.index)
```

```python
df1 = pd.read_csv('NCAA_Tourney_2002-2017.csv')
```

```python
mm_data = pd.concat([df1, df])
```
